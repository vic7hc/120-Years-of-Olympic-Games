```python
data = pd.read_csv("athlete_events.csv")
```

```python
d = {"Gold" : 3, "Silver" : 2, "Bronze" : 1}
data["Medal"] = data["Medal"].map(d)
data
```

```python
data = data.fillna(0)
data
```

```python
corr_data=data.corr()
corr_data.head()
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(),annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
data.plot(kind="scatter",x="Age",y=["Weight"])
plt.show()
```

```python
games_list = data["Games"].unique().tolist()
games_list.sort()

gamers_list = []
for i in games_list:
    gamers_list.append(len(data[data["Games"] == i]))
print(games_list,"\n")
print (gamers_list,"\n")
print(len(games_list))
```

```python
fig,ax = plt.subplots(figsize=(20,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(games_list, gamers_list)
plt.show()
```

```python
games_season = data["Games"].unique().tolist()
games_season.sort()

male = []
female = []
for i in games_season:
    data_2 = data[data["Games"]==i]
    male.append(len(data_2[data_2.Sex == "M"]["Name"].unique()))
    female.append(len(data_2[data_2.Sex == "F"]["Name"].unique()))
print("\n\nLength of male= ", len(male))
print(male)

print("\n\nLength of female= ", len(female))
print(female)
```

```python
fig,ax = plt.subplots(figsize=(15,7))
axes = fig.add_axes([0,0,1,1])

axes.plot(games_season, male, color = "blue")
axes.plot(games_season, female, color = "red")
plt.show()
```

```python
male_gamers = len(data[data["Sex"]=="M"]["Name"].unique())
female_gamers = len(data[data["Sex"]=="F"]["Name"].unique())
print(male_gamers,"\n")
print(female_gamers)
```

```python
keys = [male_gamers, female_gamers]

x_pos = [i for i, _ in enumerate(keys)]
plt.bar(x_pos, keys, color='green')

plt.xticks(x_pos, keys)
plt.show()
```
